### Making a dataset - For making a dataset of your own .

```python
import matplotlib.pyplot as plt
```

Matplotlib library is used for visulizations of the data ,by observing the value we cannot get the correct idea or nature of the dataset so we use the library.

#### Datasets can be of two types-
- Continuous values datasets - this kind of dataset used in regression 
- Categorical values datasets - this kind of dataset used in classification 

```python
from sklearn.datasets import make_regression
# make_regression is used for making regression type dataset of continuous values .
```

```python
x,y = make_regression(n_features=1,noise =5,n_samples=1000,random_state=0)
```

In the above line we declare two variables x,y x is for independent variable and y is for dependent variable
and n_features method is strictly talking about x variable not for y(Interview Question).
- noise = 5 means scattering is more (data jo hai zyada occupy kar rha hai jagah ko).

```python
x
```

```python
y
```

```python
# making a simple scatter plot graph - this graph is same as we make in school intersection of two points.
plt.scatter(x,y)
plt.xlabel('x') # label x part in the graph
plt.ylabel('y')
plt.show()
# if we reduce the noise to 0,1,2,3 then the width of the graph will be reduced.
```

```python
# making a classification dataset-
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_classes=2,n_features=5,n_samples=10000)
```

```python
# making scatter plot for classification dataset- categorical data in form of 0 and 1 just like car dataset.
plt.scatter(x[:,0],x[:,1],c=y) # x[:,0] - referring to class 0 and x[:,1] - referring to class 1 
```

```python
# blobs - we will use them in clustering just like the circles.
from sklearn.datasets import make_blobs
x,y = make_blobs(n_samples=500,centers=3,n_features=2,random_state=0)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
# Tip - Noise value should be less
from sklearn.datasets import make_circles
x,y = make_circles(n_samples=500,random_state=0,noise=0.03)
```
