```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
table = pd.DataFrame(iris.data,columns=iris.feature_names)
table.head()
```

```python
table['target'] = iris.target
table.head()
```

```python
table.shape
```

```python
setosa = table.loc[table['target']==0] # target variable having zero must be stored in setosa created object.
setosa
```

```python
virginica = table.loc[table['target']==1] # Target attributes having value 1 must be stored in virginica class.
virginica
```

```python
versicolor=table.loc[table['target']==2] # target attributes having value 2 must be stored in versicolor.
versicolor
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.plot(setosa['sepal length (cm)'],np.zeros_like(setosa['sepal length (cm)']),'s') # np.zeros_like -- jitne samples setosa mein hai toh unnte hi samples ka length hona chaiye zero array ka.
plt.plot(virginica['sepal length (cm)'],np.zeros_like(virginica['sepal length (cm)']),'o') # x , y
plt.plot(versicolor['sepal length (cm)'],np.zeros_like(versicolor['sepal length (cm)']),'d') # s --> square shape , d --> diamond , o --> circle
plt.xlabel('sepal length')
```

### Bivariate Analysis -->

```python
plt.scatter(table['sepal length (cm)'],table['petal length (cm)'],c=table['target'])
```

In the above visualization we can see that in the yellow region the sepal length and petal length both values will be more.In other words sepal and petal length have maximum values in the higher range 7.5 to 8.0 in x axis and 6-7 in y axis and the purple region describes the minimum value range of sepal length and petal length is in range of 4.5 to 5.5 in x axis and 1 to 2 in y axis.
