```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
data = load_breast_cancer()
```

```python
from sklearn.feature_selection import mutual_info_classif # important
import matplotlib.pyplot as plt
import pandas as pd
```

```python
table = pd.DataFrame(data.data,columns=data.feature_names)
table.head()
```

```python
table.shape
```

```python
imp = mutual_info_classif(data.data,data.target)
```

```python
fimp = pd.Series(imp,table.columns[0:len(table.columns)])
```

```python
fimp.sort_values(ascending=False) # The data arranged in descending order
```

```python
fimp.plot(kind='bar',color='green')
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
matrix = table.corr() # correlation can be 1 ,-1,0
```

```python
matrix # In the below matrix we can see that mean radius is highly correlated with mean radius which is present in row
# because it value is 1.00 that means that they increasing in equal manner at a particular x and y axis.
```

As we can see in the above example the mean perimeter and mean radius having value near 1 so it highly affecting the dataset values so we need to remove that feature (meam perimeter).

To get the full view of the correlation matrix we need to implement the following figure-

```python
plt.figure(figsize=(30,30))
sns.heatmap(matrix,annot=True)
```

Now to eliminate features in the dataset we are putting threshold of 0.9 (threshold value can vary).


```python
final = set()
for i in range(len(matrix.columns)):
    for j in range(i):
        if matrix.iloc[i,j] > 0.9: # i,j denoting row  and column values 
            col_name = matrix.columns[i] # we are saving column attribute and removing row attribute
            final.add(col_name)
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
sel = VarianceThreshold(threshold=0.5) #    Variance means scattering
sel.fit_transform(data.data) # if feature value is beyond threshold then it will be false and extracted from dataset otherwise it will be included.
```
