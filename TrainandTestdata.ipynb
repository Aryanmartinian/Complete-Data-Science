```python
# Another way of loading dataset by website link using pandas-
import pandas as pd 
data = pd.read_csv("https://5minutesengineering.com/wp-content/uploads/2021/05/iris.csv")
data.head()
```

```python
data.describe() # the columns which are given below are independent variables and species which we need to classify 
# is the dependent variables which depends on these independent variables.
```

```python
x = data[['sepal_length','sepal_width','petal_length','petal_width']]
y = data['species']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,shuffle=False) # by using shuffle=True keyword if cannot get the 
# training data in sequential manner we get it in a random way as 6,8,1,2 etc.
```

In testing and training of the data we should not use shuffle = False because it cannot make a good machine learning trained model.For example - In  a given dataset we have two classes , class 1 starts from index 1 to index 120 (numbered from 1 to 120) and class 2 starts from last 30 samples(numbered from last 30) so if we use shuffle keyword
or shuffle = False then it will make the training data in a sequential order from 1 to 120 and testing data from last 30 samples which belong to class 2 category . Because of this our model will be trained according to class 1 category only and cannot train on class 2 category so the model will not work properly(accuracy reduced).In classification it will give an error.

```python
print(x_train,'\n')
print(x_test,'\n')
print(y_train,'\n')
print(y_test,'\n')
```

```python
from pandas import read_csv
data1 = read_csv("age1.csv")
data1.head()
```

```python
X = data1['Age'] # independent variable  # Yeh pheli splitting hogyi hai x aur y mein(vertical split)
Y = data1['Income'] # dependent variable or outcome variable
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

In the above lines we have split the data into train and test form and divide the data into 80% training and 20% testing.random_state function is very important in the splitting of the data because when we use random_state in splitting it makes the constant split of training and testing data ,if there is no random state then the training and testing data values will change again after splitting which is not good for machine learning model.Because after making model we try to check the data on which training and testing has been done it will show a diiferent values of data in both training and testing which can reduced the accuracy of the model. ( Conclusion - For model accuracy random state method is important in testing and training data).

```python
print(X_train,'\n')
print(X_test,'\n')
print(Y_train,'\n')
print(Y_test,'\n')
```
