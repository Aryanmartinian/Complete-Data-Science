```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import datasets,metrics # from metrics module we take ROC 
from sklearn.metrics import roc_curve
import pandas as pd
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_classes=2,n_features=5,n_samples=1000,random_state=1)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
c1 = LogisticRegression()
c1.fit(X_train,Y_train)
```

```python
metrics.roc_curve(c1,X_test,Y_test)
```
