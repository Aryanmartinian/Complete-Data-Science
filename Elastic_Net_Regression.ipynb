```python
from sklearn.datasets import make_regression
x,y = make_regression(n_features=1, noise=5, n_samples=10000,random_state=2)
import matplotlib.pyplot as plt
plt.scatter(x,y,s=3)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import ElasticNet
reg = ElasticNet()
reg.fit(X_train,y_train) # Model has been trained on the Training data.
```

```python
reg.predict(X_test)
```

```python
reg.score(X_test, y_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred = reg.predict(x)
```

```python
plt.scatter(x,y,label='Actual',s=3)
plt.scatter(x,y_pred,label='Predicted',s=3)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
reg = ElasticNet(alpha=2)
reg.fit(X_train,y_train)
reg.score(X_test, y_test)
```

```python
y_pred = reg.predict(x)
print(reg.coef_)
```

```python
plt.scatter(x,y,label='Actual',s=3)
plt.scatter(x,y_pred,label='Predicted',s=3)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
