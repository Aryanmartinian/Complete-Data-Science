```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("job.csv")
df.head(10)
```

- O --> means not successful
- 1 --> means successful

```python
x = df[['Years']]
y = df['result']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression() # Creating a model object
model.fit(X_train,Y_train)
```

```python
print(X_test)
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
Y_test
```

```python
model.score(X_test,Y_test) # Model accuracy score jo hai internally predict kiya aur score algorithm run kiya.
```

```python
import seaborn as sns
sns.regplot(x=X_test, y=Y_test, logistic=True) # Sigma function 
```
