```python
import pandas as pd
example = pd.read_csv("age1.csv")
example
```

```python
example['Income'].notnull() # Returns boolean value if there is null then false and if not null then it is true.
```

```python
example[example["Income"].notnull()] # to get the exact location of not null value in the dataset
```

```python
example[example['Income'].isnull()] # giving me the exact location of null values.
```

```python
example['Income'].fillna(example['Income'].max()) # filling the null values with the maximum income price that the 
# income dataset has.
```

```python
example['Income'] = example['Income'].fillna(example['Income'].std())
example
```

```python
example['Income'] = example['Income'].fillna(example['Income'].min())
example
```

Now we can see that in the above three queries the values are not changing because in the first queries itself we have use fillna method which is fill the null values with the max value that the income method has so after this query the null values has been acquired , after that max query next two queries will cannot change it because the null values places are already filled by the first query.
