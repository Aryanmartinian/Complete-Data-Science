```python
from sklearn.datasets import make_blobs,make_moons
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x,y = make_blobs(n_features=2,n_samples=5000,centers=4)
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(x)
```

```python
predict = kmeans.predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=predict)
```

In the above diagram the data is clusterized.

Below is the demonstrating at which the clustering does not work properly(K-Means does not support)|

```python
x,y = make_moons(n_samples=1000,noise=0.06)
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(x)
```

```python
predict = kmeans.predict(x)
```

```python
plt.scatter(x[:,0], x[:,1],c=predict)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ag1 = AgglomerativeClustering(linkage='single')
```

```python
ag1.fit(x)
```

```python
ag1.labels_
```

```python
plt.scatter(x[:,0], x[:,1],c=ag1.labels_)
```
