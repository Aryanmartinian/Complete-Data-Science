### TfidVectorizer --

```python
# TfidVectorizer - Converting the word into numerical values vectors so that it can receive or given to the machine Learning model.
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
A1 = "hello and welcome dosto"
A2 = 'shri love NLP'
A3 = 'shri is good boy'
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
vectors = vectorizer.fit_transform([A1,A2,A3])
feature_names = vectorizer.get_feature_names_out()
dense = vectors.todense()
result = pd.DataFrame(dense,columns=feature_names)
print(result)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
import pandas as pd
```

```python
dataset = pd.read_csv('finalreview.csv')
dataset
```

```python
# Basic Pre-Processing -
from nltk.corpus import stopwords
import re
corpus = []
for i in range(0, 29):
    review = re.sub('[^a-zA-Z]', ' ', dataset['review'][i])
    review = review.lower()
    review = review.split() 
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Another step of Basic Pre-processing into tokens -(Best way)
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

sentences = []
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        sentences.append(simple_preprocess(sent))
```

```python
import gensim # Making tokens using gensim
model = gensim.models.Word2Vec(sentences,window=3,min_count=1)
```

```python
model.wv.index_to_key
```

```python
from gensim.models import KeyedVectors
word_vectors = model.wv
word_vectors.save("word2vec.wordvectors")

wv = KeyedVectors.load("word2vec.wordvectors") # saving the word vector file
vector = wv['worst']
```

```python
import numpy as np
def sentence_vector(sent):
    doc = [word for word in sent.split() if word in model.wv.index_to_key]
    return np.mean(model.wv[doc], axis=0)
```

```python
sentence_vector(dataset['review'].values[20]) # index number 20 is send to sentence_vector
```

```python
from tqdm import tqdm # This library is used to check the progress of the task
```

```python
final = []
for s in tqdm(dataset['review'].values):
    final.append(sentence_vector(s))
```
