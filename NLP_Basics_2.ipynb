### TfidVectorizer --

```python
# TfidVectorizer - Converting the word into numerical values vectors so that it can receive or given to the machine Learning model.
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
A1 = "hello and welcome dosto"
A2 = 'shri love NLP'
A3 = 'shri is good boy'
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
vectors = vectorizer.fit_transform([A1,A2,A3])
feature_names = vectorizer.get_feature_names_out()
dense = vectors.todense()
result = pd.DataFrame(dense,columns=feature_names)
print(result)
```
