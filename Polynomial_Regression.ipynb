```python
import matplotlib.pyplot as plt
import numpy as np

X = np.array([2, 3, 4, 5, 6, 7, 8, 9])
Y = np.array([4, 6, 8, 6, 5, 4, 7, 8])

plt.scatter(X, Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state=0)
```

```python
X_train=X_train.reshape(-1, 1)
X_train
```

```python
X_test=X_test.reshape(-1, 1)
X_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,X_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
plt.scatter(X_test,y_test,label='Actual')
plt.plot(X_test,y_pred,label='Predicted')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

Now in the above diagram you can see that the data points are very far away from the straight line so the model accuracy is not perfect and linear regression is not working. The reason of linear Regression is not working is that the data which we have made is very scatter.(faalaa hua hai)

### Polynomial Regression - 

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2) # in this function we give reshape data .
poly_x = poly.fit_transform(X.reshape(-1,1)) # it converts the data into degree 2 ( quadratic mein data transform ho chuka hai)
model1 = LinearRegression()
model1.fit(poly_x,Y)
y_pred = model1.predict(poly_x)
plt.scatter(X,Y)
plt.plot(X,y_pred,c='red')
poly_x # b0+b1x1+b2x2
```

```python
model1.score(poly_x, Y)
```
