Robust Regression is robust to outliers that means if there are outliers in the data it will cannot affect the robust regression performance.

Outliers - Suppose in a class if there 6 students total and 5 of them scored 32% in a exam and 1 student has score more than 90% then if a teacher wants to find out that which class students need extra class they take out mean(average) of all classes and decide if the average percentage is below 35% then only that class will be given extra class and above one's don't need any extra class , so in this scenario while taking out the mean the above 90% score increased the mean and crossed the threshold mark of 35 % then the teacher cannot get the correct idea of whom students need extra class . In this the outlier was that 90% student.Basically outlier mean that the value which is very much different from the other values present in the dataset.

```python
from sklearn.datasets import make_regression
import numpy as np 
from sklearn.linear_model import LinearRegression,RANSACRegressor # Robust Regression
import matplotlib.pyplot as plt
```

```python
x,y = make_regression(n_samples=200000,n_features=1,noise=10,random_state=0)
outliers = 30000 # defining the number of outliers
np.random.seed(2) # Random Numbers Normalize - same random numbers baar baar generate karege .
y[:outliers] = 600 + np.random.normal(size=(outliers)) # outliers ki value 600 ke around saturate hojayegi(shown in the graph)
```

```python
sample = np.random.normal(size=(outliers))
```

```python
plt.scatter(x,y,s=3)
```

The code plt.scatter(x, y, s=3) creates a scatter plot using the matplotlib library in Python. Here's what each parameter does:

x: This is an array-like object containing the x-coordinates of the data points to be plotted.
y: This is an array-like object containing the y-coordinates of the data points to be plotted.
s: This parameter sets the size of the markers (data points) in the scatter plot. In this case, s=3 means each marker will have a size of 3.

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
lr = LinearRegression()
lr.fit(X_train,Y_train) # Train the model on training data.
```

```python
rr = RANSACRegressor()
rr.fit(X_train,Y_train)
```

```python
rr_pred = rr.predict(X_test)
```

```python
lr_pred = lr.predict(X_test)
```

```python
plt.scatter(X_test,Y_test,label='Actual data',c="green",s=2)
plt.plot(X_test,rr_pred,label='RANSACRegressor',c="red")
plt.plot(X_test,lr_pred,label='LinearRegression',c="black")
plt.legend()
```

In the above diagram we can see that because of outliers linear regression line shifts from the actual part and deviate little but the ransac regression line does not shift it stays robust i.e does not shift.
