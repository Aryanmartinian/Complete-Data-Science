```python
import pandas as pd
df = pd.read_csv("age2.csv")
df.head(6)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Age'],df['Income'])
plt.xlabel('Age')
plt.ylabel("Income")
plt.show()
```

```python
plt.scatter(df['Degrees'],df['Income'])
plt.xlabel('Degrees')
plt.ylabel('Income')
plt.show()
```

```python
x = df[['Age', 'Degrees']]
y = df['Income']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=0) 
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
```

```python
clf.fit(X_train,Y_train) # Model trained on training data
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test,Y_test) # Model accuracy is 97%.
```

```python
clf.coef_ # negative kyu hai because it is slope
```

```python
clf.intercept_
```

```python
y_pred = clf.predict(x)
y_pred
```

```python
import seaborn as sns
sns.regplot(x=y,y=y_pred,ci=None) # x axis par actual income hai jo dataset mein di rakhi hai aur y axis par predicted values hai.
```

Upar diagram mein blue line jo hai woh actual predicted value hai aur blue circles jo hai woh actual values hai.

### Task - Make your own regression data and build a linear regression model and predict the accuracy-->


```python
from sklearn.datasets import make_regression
x,y = make_regression(n_features=4,noise=4,n_samples=10000) # My data has been created.
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train) # Linear Regression Model train hogya hai.
```

```python
regressor.predict(X_test)
```

```python
regressor.score(X_test,Y_test)
```
